```python
print('TOP 10 Happiest Countries (2019):')
print(df.nlargest(10, 'Happiness Rank')[['Country','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)']].to_string(index=False))

print('\nBOTTOM 10 Least Happy Countries (2019):')
print(df.nsmallest(10, 'Happiness Rank')[['Country','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)']].to_string(index=False))
```

```python
df.sort_values(by="Happiness Score", ascending=False) \
  .head(10)[["Country", "Happiness Score"]] \
  .style.background_gradient(cmap="viridis")
```

```python
df["Region"].value_counts().to_frame().style.background_gradient(cmap="twilight")
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["Happiness Score"], kde=True)
plt.title("Distribution of Happiness Score", fontsize=16)
plt.xlabel("Happiness Score", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.show()
```

```python
df["Region"].value_counts().plot(kind="barh", figsize=(13,6),color='lightblue')
plt.title("Number of Countries per Region")
plt.xlabel("Number of Countries")
plt.ylabel("Region")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x="Health (Life Expectancy)", y="Happiness Score", data=df)
plt.title("Health vs Happiness Scpre", fontsize=16)
plt.xlabel("Health (Life Expectancy)", fontsize=14)
plt.ylabel("Happiness Score", fontsize=14)
plt.show()
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Happiness Dataset Features", fontsize=16)
plt.show()
```
